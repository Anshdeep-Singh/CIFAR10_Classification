```
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

# Loading the Model obtained from running the model training file, follow the Readme instructions if any of the cell not executing.
model = tf.keras.models.load_model('CIFAR10_.h5')

# Storing the the training data and testing data in x_train, y_train, x_test and y_test 
(p_x_train , p_y_train) , (p_x_test , p_y_test) = cifar10.load_data()
```

```
# The summary of the training model
model.summary()
```

```
# Running this cell will take a while compared to other cells
# Normalizing the values and making preditions
p_x_test = p_x_test/255.0 # normalizing the x_test values and x_train values
p_x_train = p_x_train/255.0
predictions = np.argmax(model.predict(p_x_test), axis=-1)

# Prints the Classsification and Confusion Report
print(classification_report(p_y_test ,predictions))
print(confusion_matrix(p_y_test ,predictions))
```

```
# Creating A Dictionary that stores the names of the figures corresponding to their indices for simpilicity in understanding the output
image = {0: 'Airplane', 1: 'Automobile',
         2: 'Bird', 3: 'Cat', 4: 'Deer',
         5: 'Dog', 6: 'Frog', 7: 'Horse',
         8: 'Ship', 9: 'Truck'} 
```

```
# To Test the model Change the value of X to any value between 0 and 9999
X=12 # Replace X with any integer value between 0 to 9999
test_model = p_x_test[X]
plt.matshow(p_x_test[X])
print("REAL LABEL:",image[int(p_y_test[X])])

test_model = test_model.reshape(1, 32, 32,3)    # Reshaping the image for the model
test_model = np.argmax(model.predict(test_model), axis=-1)   # Prediction process
print("PREDICTION :", image[test_model[0]])      # printing the name of the image predicted by the model 


```

```
flag1 =0
flag2=0
faults = []
for x in range(0,15):
  test_model = p_x_test[x]
  test_model = test_model.reshape(1, 32, 32,3)
  test_model = np.argmax(model.predict(test_model), axis=-1)

  if test_model[0]==p_y_test[x]:
    flag1 = flag1 +1
    
  else:
    faults.append(x)
    flag2 = flag2 + 1

print(faults)
print("Correct:",flag1," Not Correct:", flag2)
```
